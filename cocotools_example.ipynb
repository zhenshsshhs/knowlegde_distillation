```python
from pycocotools.coco import COCO
import skimage.io as io
import matplotlib.pyplot as plt
```

```python
annFile = '/data1/zhengshuaijie/data/annotations/instances_val2017.json'
coco = COCO(annotation_file=annFile)
img = coco.loadImgs([324158])[0]
I = io.imread('/data1/zhengshuaijie/data/val2017/'+img['file_name'])
plt.imshow(I)
plt.show()
```

```python
plt.imshow(I)
annIds = coco.getAnnIds(imgIds=img['id'])
anns = coco.loadAnns(annIds)
coco.showAnns(anns)
```
